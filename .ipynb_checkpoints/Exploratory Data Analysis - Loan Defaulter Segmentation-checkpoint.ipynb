# Feature Engineering

### Missing Imputation

```python
app_score_col_rmvd.isna().sum().sort_values()/app_score_col_rmvd.shape[0]
```

```python
app_score_col_rmvd['CNT_FAM_MEMBERS'] = app_score_col_rmvd['CNT_FAM_MEMBERS'].fillna((app_score_col_rmvd['CNT_FAM_MEMBERS'].mode()[0]))
```

```python
app_score_col_rmvd['CNT_FAM_MEMBERS'].isna().sum()
```

```python
app_score_col_rmvd['OCCUPATION_TYPE'] = app_score_col_rmvd['OCCUPATION_TYPE'].fillna(app_score_col_rmvd['OCCUPATION_TYPE'].mode()[0])
```

```python
app_score_col_rmvd['NAME_TYPE_SUITE'] = app_score_col_rmvd['NAME_TYPE_SUITE'].fillna(app_score_col_rmvd['NAME_TYPE_SUITE'].mode()[0])
```

```python
app_score_col_rmvd['NAME_TYPE_SUITE'].isna().sum()
```

```python
app_score_col_rmvd['AMT_ANNUITY'] = app_score_col_rmvd['AMT_ANNUITY'].fillna(app_score_col_rmvd['AMT_ANNUITY'].mean())
```

```python
app_score_col_rmvd['AMT_ANNUITY'].isna().sum()
```

```python
amt_req_col = []

for col in app_score_col_rmvd.columns:
    if col.startswith("AMT_REQ_CREDIT_BUREAU"):
        amt_req_col.append(col)
amt_req_col
```

```python
for col in amt_req_col:
    app_score_col_rmvd[col] = app_score_col_rmvd[col].fillna(app_score_col_rmvd[col].median())
```

```python
app_score_col_rmvd['AMT_GOODS_PRICE'].agg(['min','max','median', 'mean'])
```

```python
app_score_col_rmvd['AMT_GOODS_PRICE'] = app_score_col_rmvd['AMT_GOODS_PRICE'].fillna(app_score_col_rmvd['AMT_GOODS_PRICE'].median())
```

```python
app_score_col_rmvd['AMT_GOODS_PRICE'].isna().sum()
```

### Value Modification

```python
days_col = []

for col in app_score_col_rmvd.columns:
    if col.startswith('DAYS'):
        days_col.append(col)
days_col
```

```python
for col in days_col:
    app_score_col_rmvd[col] = abs(app_score_col_rmvd[col])
```

```python
app_score_col_rmvd.head()
```

NAME_CONTRACT_TYPE
* Most of the customers have taken cash loan.
* Customers who have takesh cash loans are less likely to default.

CODE_GENDER
* Most of the loans have been taken by female.
* Default rate for females are just approximately 7% which is safer.

NAME_TYPE_SUITE
* Unaccompanied people had taken most of the loans and the default rate if approximately 8% which is still safer.

NAME_INCOME_TYPE
* The safest segments are Working , commercial associates and pensioners.

NAME_EDUCATION_TYPE
* Given that Higher/secondary specials are taking most number of loans they have the highest default rates.
* Higher education is the safest segment to give the loan with a default rate of less than 5%.

NAME_FAMILY_STATUS
* Even though Married people are taking more no of loans, they are safe to target as they have the mediocre default rate as compared to others.

NAME_HOUSING_TYPE
* People having house/apartment are safe to give the loan with default rate of approximately 8%.

OCCUPATION_TYPE
* Laborers are the majority occupation segment who takes most number of loans and have mediocre default rate.
* Low-skill Laborers are highest defaulters.
* Accountants are less defaulters and might be safer segment to target.
* Core Staff and manager are safe to target who has approximately 10% default rate.

ORGANIZATION_TYPE
* Transport Type 3 has the highest default rate.
* Others, Buiness Type 3, Self employeed have good default rate and can target them for the loans.

## CONCLUSIONS/INSIGHTS

* Banks should target the customers- 
    * Low Income customers (Having income below 1 million) can be targeted.
    * Customers having low income 
    * Working in Others, Buisness, Entity Type 3, Self Employeed org. type.
    * Working as Accountants, Core staff, Managers, and Laborers.
    * Having House/Apartments and are married and having children not more than 5.
    * Highly educated people should be targeted.
    * Female customers should be preffered more.


* Unaccompanied people can be safer.


* Amount Segment reccommended- 
    * The credit amount should now be more than 1 million.
    * Annuity can be made of 50K (depending on the eligibility)
    * Income bracket can be below 1 million.


* 80-90% of customers whose previous applications were refused/cancelled are repayers. Bank can do the analysis and can consider to give loans to these segments.



* Organization Transport type 3 should be avoided.
* Low-skill Laborers and drivers should be avoided.
* Offers previously unused and high income customers should be avoided.
